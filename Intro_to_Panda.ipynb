## <font color='teal'> Pandas</font>

#### Pandas stands for Panel Data and is the core library for data manipulation and data analysis. It consists of single and Multi-dimensional data structure for data manipulation.Single dimensional in pandas as a series object and Multi-dimensional as a Data-Frame.

#### Pandas Series Object - <font size=2>Series Object is one-dimensional labeled array.</font>

```python
import pandas as pd
s1 = pd.Series([1,2,3,4,5])
s1
```

```python
type(s1)
```

### Changing Index

```python
s2 = pd.Series([1,2,3,4,5])
s2
```

```python
s2 = pd.Series([1,2,3,4,5], index=['a','b','c','d','e'])
s2
```

### Series Object from Dictionary <font size = 2>(We can create a series object from a dictionary)</font>

```python
s3 = pd.Series({'a':100,'b':200,'c':300})
s3
```

### Changing Index Position

```python
s4 = pd.Series({'a':100,'b':200,'c':300}, index=['a','c','b','d'])
s4
```

### Extracting Individual Elements

```python
s5 = pd.Series([10,20,30,40,50,60,70,80])
s5
```

```python
s5[6]
```

```python
s5[:6]
```

```python
s5[-5:]
```

### DataFrame InBuilt Functions
head() <br>
tail() <br>
Shape() <br>
describe() <br>

```python
gdp = pd.read_csv('GDP.csv')
```

```python
gdp.head()                 # Print First Five records of file
```

```python
gdp.tail()                 # Print Last Five records of file
```

```python
gdp.shape                 # To know the how many rows and columns are in the file
```

```python
gdp.describe()            # Print the mean,standard deviation etc.
```

### iloc[] help to extract rows and columns with respect to that index 

```python
gdp.iloc[0:5,-5:]       #Extract the first 5 rows and last 5 columns
```

#### loc[] for columns by passing name of column

```python
gdp.loc[:,["Country ","2017","2018","2019"]]
```

### Droping Columns

```python
gdp.drop('2019',axis = 1)    # Axis=1 represent column
```

### Dropping Rows

```python
gdp.drop([1,2],axis = 0)
```
